### 2.1 My Motivation and Goal
When I was assigned this task, I was overwhelmed by the numerous possibilities of using LSTM. After working through an example from the book in which an LSTM generated text based on Nietzsche's writings, I was undoubtedly amazed. Completely new texts sequences were generated by a trained LSTM - and this literally happened from the code that came from my laptop! So cool. But on the other hand I realized that the original input from Nietzsche was kind of difficult to understand. So the outcomes were also somehow abstruse. So I was thinking, can a LSTM maybe generate a meaningful text? Could it read between the lines? Could it maybe find a specific concept, which is not explicitly apparent in the given input sequences? <br>

Stereotypes came to my mind. Paradoxes came to my mind. Concepts like Good and Bad came to my mind. This lead me to the choice of my final topic: Fairy tales. More precisely - the fairy tales of the Grimm brothers. We all know the famous fairy tales of Hansel and Gretel, the Little Red Riding Hood, Snow White, Rapunzel, Sleeping Beauty, Rumpelstiltskin and so on. When I was little I loved these fairytales because to me they were simple and clear: the good always wins and the bad gets punished. Easy, right? Unfortunately, it's not as simple as it looks and it is also not the case in the real world. As I grew up, I eventually realized that these fairy tales are somehow cruel, sexist and also violent. As an example, lets look at the last two sentences from the the Snow White fairy tale: 

> [...] But iron slippers had already been placed over coal fires and were carried in with tongs and placed in front of her. Then she had to step into the red-hot shoes and dance until she fell dead to the ground.

I'm sorry, what?! That is how Snow Whites step mother is treated in the end of the fairy tale. Well, welcome to the year 1812, where this was apparently okay... As a child I did not recognize this cruelty. The Good won and the Bad lost, that was the most important thing. Somehow the end justified the means. Fortunately, as adults we are all now able to critically reflect on these fairy tales. So, why not go one step further and include an NN in this topic? That is what I will try to do here. The basic question that I want to answer in this tutorial is the following: <br>

**Can a trained LSTM model, which receives the Grimm fairy tales as input, recognize its stereotypical, sexist and/or violent characteristics?**
